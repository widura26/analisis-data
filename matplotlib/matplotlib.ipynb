### Matplotlib Pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Draw a line in a diagram from position (0,0) to position (6,250)
xpoints = np.array([0,6])
ypoints = np.array([0,250])

plt.plot(xpoints, ypoints)
plt.show()
```

### Matplotlib Plotting

```python
# Plotting x and y points
xpoints1 = np.array([1,2,8])
ypoints1 = np.array([3,5,10])

plt.plot(xpoints1, ypoints1)
plt.show()
```

```python
# Plotting Without Line
xpoints2 = np.array([1,8])
ypoints2 = np.array([3,10])

plt.plot(xpoints2, ypoints2, 'o')
plt.show()
```

```python
# Multiple Points

xpoints3 = np.array([1,2,6,8])
ypoints3 = np.array([3,8,1,10])

plt.plot(xpoints3, ypoints3)
plt.show()
```

```python
# Default x-points
ypoints4 = np.array([3,8,1,10,5,7])
plt.plot(ypoints4)
plt.show()
```

### Matplotlib Markers

```python
# Markers
ypoints5 = np.array([3,8,1,10])
plt.plot(ypoints5, marker = 'o')
plt.show()
```

```python
# Format Strings (fmt)
ypoints6 = np.array([3,8,1,10])
plt.plot(ypoints6, 'o:r')
plt.show()
```
